```python
import spacy
import re
from cairosvg import svg2png

nlp = spacy.load("en_core_web_trf")
sent = "CALFIRE and the Winters FD lead a DS Assessment Training hosted by Yolo County FSC and Solano RCD."
doc = nlp(sent)

options = {
    "color": "black",
    "compact": True,
    "collapse_phrases": True,
    "fine_grained": True,
    "font": "Arial",
   # "word_spacing": 25,
    "distance": 125,  # smaller = narrower image (default 175)
}

# Render to SVG
svg = spacy.displacy.render(doc, style="dep", jupyter=False, options=options)

# Shift arc labels below the arc lines
dy = 12  # adjust as needed
svg_adjusted = re.sub(
    r'(<textPath[^>]*displacy-label[^>]*)',
    lambda m: m.group(1) + f' dy="{dy}"',
    svg
)

# Save as high-res PNG
svg2png(bytestring=svg_adjusted.encode(), write_to="../../figures/dependency_parse.png", scale=3.0)
print("Saved dependency_parse.png")
```
