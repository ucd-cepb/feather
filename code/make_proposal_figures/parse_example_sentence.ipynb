```python
import spacy
nlp = spacy.load("en_core_web_sm")
sent = "CALFIRE and the Winters_FD lead a DS_Assessment_Training hosted by Yolo County FSC, Green Valley FSC and Solano RCD."
doc = nlp(sent)
```

```python
spacy.displacy.serve(doc, style="ent",auto_select_port=True)
```

```python

options = {"compact": True, "color": "black","collapse_phrases": True,"fine_grained": True,"font": "Times"}
spacy.displacy.serve(doc, style="dep",options = options,auto_select_port=True)
```

```python
# Render the dependency parse as SVG
svg = spacy.displacy.render(doc, style="dep", jupyter=False,options = options)

# Save the SVG to a file
with open("dependency_parse.png", "w") as f:
    f.write(svg)
```
